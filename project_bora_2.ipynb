```python
import geopandas as gpd
from geopy.geocoders import Nominatim
import numpy as np

# 도시와 국가별 고객 수 집계
customer_distribution = data.groupby(['Order City', 'Order Country']).size().reset_index(name='Customer Count')

# 지오로케이터 초기화
geolocator = Nominatim(user_agent="geoapiExercises")

# 위도와 경도를 가져오는 함수
def get_lat_long(city, country):
    try:
        location = geolocator.geocode(f"{city}, {country}")
        if location:
            return location.latitude, location.longitude
        else:
            return np.nan, np.nan
    except:
        return np.nan, np.nan

# 각 도시와 국가에 대한 위도 및 경도 가져오기
customer_distribution[['Latitude', 'Longitude']] = customer_distribution.apply(
    lambda row: get_lat_long(row['Order City'], row['Order Country']), axis=1, result_type='expand'
)

# 좌표를 가져오지 못한 행 제거
customer_distribution.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# 세계 지도를 기반으로 산점도 그리기
fig, ax = plt.subplots(figsize=(15, 10))

# 자연 지구 데이터 경로를 지정하세요 
world = gpd.read_file('C:/Users/dlaqh/OneDrive/FILE/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp')

# 세계 지도 그리기
world.plot(ax=ax, color='lightgray')

# 고객 위치 산점도 그리기
plt.scatter(
    customer_distribution['Longitude'], 
    customer_distribution['Latitude'], 
    s=customer_distribution['Customer Count'] / 10,  # 고객 수에 따라 점의 크기 조정
    alpha=0.6, 
    color='red',
    edgecolors="black"
)

# 제목 및 레이블 추가
plt.title('도시 및 국가별 고객 지리적 분포', fontsize=15)
plt.xlabel('경도', fontsize=12)
plt.ylabel('위도', fontsize=12)

# 플롯 표시
plt.show()
```

```python
# RFM 분석
rfm_df = data[['Customer Id', 'order date (DateOrders)', 'Sales per customer']]
rfm_df = rfm_df.rename(columns={'Customer Id': 'CustomerID', 'order date (DateOrders)': 'OrderDate', 'Sales per customer': 'Sales'})

# 'OrderDate' 컬럼을 날짜 형식으로 변환
rfm_df['OrderDate'] = pd.to_datetime(rfm_df['OrderDate'], errors='coerce')

# 기준일을 데이터의 마지막 구매일 이후로 설정
analysis_date = rfm_df['OrderDate'].max() + pd.Timedelta(days=1)  # 하루를 추가한 날짜로 설정

# RFM 분석을 위한 Recency, Frequency, Monetary 값 계산
rfm = rfm_df.groupby('CustomerID').agg(
    Recency=('OrderDate', lambda x: (analysis_date - x.max()).days),  # Recency: 가장 최근 구매일로부터의 일수
    Frequency=('OrderDate', 'count'),                                # Frequency: 총 구매 횟수
    Monetary=('Sales', 'sum')                                        # Monetary: 총 구매 금액
).reset_index()

# Recency, Frequency, Monetary 값을 적절한 구간으로 나누기 위해 cut 함수를 사용합니다.
rfm['R_Score'] = pd.cut(rfm['Recency'], bins=5, labels=[5, 4, 3, 2, 1])
rfm['F_Score'] = pd.cut(rfm['Frequency'], bins=5, labels=[1, 2, 3, 4, 5])
rfm['M_Score'] = pd.cut(rfm['Monetary'], bins=5, labels=[1, 2, 3, 4, 5])

# 최종 RFM 점수를 계산
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# 고객 세그먼트 분류 (예: VIP, 일반 고객 등)
def segment_customer(df):
    if df['RFM_Score'] >= '555':
        return 'VIP'
    elif df['RFM_Score'] >= '355':
        return 'Loyal'
    elif df['RFM_Score'] >= '155':
        return 'Potential'
    else:
        return 'Regular'

rfm['Customer Segment'] = rfm.apply(segment_customer, axis=1)

# RFM 분석 결과를 고객 세그먼트별로 집계하여 표시
rfm_summary = rfm.groupby('Customer Segment').agg(
    Customer_Count=('CustomerID', 'nunique'),
    Average_Recency=('Recency', 'mean'),
    Average_Frequency=('Frequency', 'mean'),
    Average_Monetary=('Monetary', 'mean')
).reset_index()

# RFM 분석 요약 결과 출력
print("RFM Analysis Summary:")
print(rfm_summary)
```

```python

# rfm_summary 데이터 구조를 확인하고 시각화를 준비
rfm_summary = pd.DataFrame({
    'Customer Segment': ['VIP', 'Loyal', 'Potential', 'Regular'],
    'Customer_Count': [rfm['Customer Segment'].value_counts().get('VIP', 0),
                       rfm['Customer Segment'].value_counts().get('Loyal', 0),
                       rfm['Customer Segment'].value_counts().get('Potential', 0),
                       rfm['Customer Segment'].value_counts().get('Regular', 0)],
    'Average_Recency': rfm.groupby('Customer Segment')['Recency'].mean().tolist(),
    'Average_Frequency': rfm.groupby('Customer Segment')['Frequency'].mean().tolist(),
    'Average_Monetary': rfm.groupby('Customer Segment')['Monetary'].mean().tolist()
})

# 시각화: 고객 세그먼트별 고객 수
plt.figure(figsize=(10, 6))
plt.bar(rfm_summary['Customer Segment'], rfm_summary['Customer_Count'], color='skyblue')
plt.title('고객 세그먼트별 고객 수')
plt.xlabel('고객 세그먼트')
plt.ylabel('고객 수')
plt.show()

# 시각화: 고객 세그먼트별 평균 Recency, Frequency, Monetary 비교
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=False)

# 평균 Recency 시각화
axes[0].bar(rfm_summary['Customer Segment'], rfm_summary['Average_Recency'], color='lightcoral')
axes[0].set_title('세그먼트별 평균 Recency')
axes[0].set_xlabel('고객 세그먼트')
axes[0].set_ylabel('평균 Recency')

# 평균 Frequency 시각화
axes[1].bar(rfm_summary['Customer Segment'], rfm_summary['Average_Frequency'], color='lightgreen')
axes[1].set_title('세그먼트별 평균 Frequency')
axes[1].set_xlabel('고객 세그먼트')
axes[1].set_ylabel('평균 Frequency')

# 평균 Monetary 시각화
axes[2].bar(rfm_summary['Customer Segment'], rfm_summary['Average_Monetary'], color='lightblue')
axes[2].set_title('세그먼트별 평균 Monetary')
axes[2].set_xlabel('고객 세그먼트')
axes[2].set_ylabel('평균 Monetary')

plt.tight_layout()
plt.show()
```

```python
# 결측치를 제거하고 분석에 필요한 데이터 추출
channel_analysis = data.dropna(subset=['Customer Segment'])

# 고객 유형별  총 매출, 평균 구매 금액, 구매 빈도, 고객 수 집계
channel_behavior_summary = channel_analysis.groupby('Customer Segment').agg(
    Total_Sales=('Sales per customer', 'sum'),       # 총 매출
    Purchase_Frequency=('Order Id', 'count'),        # 구매 빈도
    Total_Customers=('Customer Id', 'nunique')    # 고유 고객 수
).reset_index()

# 고객 유형별 고객 행동 분석 결과 출력
print("판매 채널별 고객 행동 분석 결과:")
print(channel_behavior_summary.sort_values(by='Total_Sales', ascending=False))

# 시각화: 판매 채널별 총 매출
plt.figure(figsize=(14, 6))
plt.bar(channel_behavior_summary['Customer Segment'], channel_behavior_summary['Total_Sales'], color='lightblue')
plt.xlabel('Customer Segment (Sales Channel)')
plt.ylabel('총 매출')
plt.title('판매 채널별 총 매출')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시각화: 판매 채널별 구매 빈도
plt.figure(figsize=(14, 6))
plt.bar(channel_behavior_summary['Customer Segment'], channel_behavior_summary['Purchase_Frequency'], color='lightcoral')
plt.xlabel('Customer Segment (Sales Channel)')
plt.ylabel('구매 빈도')
plt.title('판매 채널별 구매 빈도')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시각화: 판매 채널별 고객 수
plt.figure(figsize=(14, 6))
plt.bar(channel_behavior_summary['Customer Segment'], channel_behavior_summary['Total_Customers'], color='lightpink')
plt.xlabel('Customer Segment (Sales Channel)')
plt.ylabel('고유 고객 수')
plt.title('판매 채널별 고객 수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# 버블 차트 생성
plt.figure(figsize=(10, 6))
bubble = plt.scatter(channel_behavior_summary['Purchase_Frequency'], 
                     channel_behavior_summary['Total_Sales'], 
                     s=channel_behavior_summary['Total_Customers'] * 10, 
                     alpha=0.5, 
                     c=channel_behavior_summary['Total_Sales'], # 총 매출에 따라 색상 변경
                     cmap='viridis', 
                     edgecolors='black', 
                     linewidth=0.5)

# 그래프에 텍스트 레이블 추가
for i in range(channel_behavior_summary.shape[0]):
    plt.text(channel_behavior_summary['Purchase_Frequency'][i], 
             channel_behavior_summary['Total_Sales'][i], 
             channel_behavior_summary['Customer Segment'][i], fontsize=9, ha='right')

# 그래프 설정
plt.title('Sales vs Purchase Frequency by Customer Segment')
plt.xlabel('Purchase Frequency')
plt.ylabel('Total Sales')
plt.colorbar(bubble, label='Total Sales')
plt.grid(True, linestyle='--', alpha=0.7)

# 버블 크기 
for size in [30, 60, 100]:
    plt.scatter([], [], s=size * 10, c='gray', alpha=0.5, label=str(size) + ' Customers')
plt.legend(scatterpoints=1, frameon=True, labelspacing=1, title='Unique Customers')

# 그래프 표시
plt.show()
```
